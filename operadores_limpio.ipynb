# Lab - Operadores: Ticket simple (sin funciones)

Objetivo: practicar operadores aritm√©ticos, de residuo y exponente,
y verificar comparaciones y operadores l√≥gicos como expresiones.

üëâ Complete los TODO y ejecute las celdas en orden.

CELDA (Code) ‚Äî Setup (NO tocar)

```python
# =========================
# SETUP (NO MODIFICAR)
# =========================

# Productos (valores fijos)
precio_cafe = 8.00
qty_cafe = 3

precio_galletas = 2.50
qty_galletas = 7

precio_te = 3.25
qty_te = 2

# Regla 2x1: aplica solo a algunos productos
cafe_aplica_2x1 = True
galletas_aplica_2x1 = True
te_aplica_2x1 = False

# Impuesto (ejemplo)
tax_rate = 0.13

# Variables para practicar comparaci√≥n y l√≥gicos (sin if)
es_vip = True
vip_threshold = 100.00
codigo_cupon = "PROMO5"
```

## Paso 1 ‚Äî Subtotales (aritm√©ticos)

Calcule el subtotal de cada producto (price * qty) y luego el subtotal general.

```python
# Subtotal por producto
subtotal_cafe = precio_cafe*qty_cafe
subtotal_galletas = precio_galletas*qty_galletas
subtotal_te = precio_te*qty_te
```

```python
# Subtotal general
subtotal = subtotal_cafe + subtotal_galletas + subtotal_te
print("Subtotal:", subtotal)
```

## Paso 2 ‚Äî 2x1 usando divisi√≥n entera `//`

Regla: por cada 2 unidades, se regala 1.
Ejemplos:
- qty=3 ‚Üí 3//2 = 1 unidad gratis
- qty=7 ‚Üí 7//2 = 3 unidades gratis

üëâ Calcule las unidades gratis y el descuento total del 2x1.
Nota: si un producto NO aplica 2x1, su 'gratis' debe ser 0.

```python
# Unidades gratis (use //)
gratis_cafe = qty_cafe//2
gratis_galletas = qty_galletas//2
gratis_te = 0  # NO aplica 2x1, por eso es 0
```

```python
# Descuento total 2x1
descuento_2x1 = (gratis_cafe*precio_cafe)+(gratis_galletas*precio_galletas)+(gratis_te*precio_te)
```

```python
print("Gratis caf√©:", gratis_cafe, "| Gratis galletas:", gratis_galletas, "| Descuento 2x1:", descuento_2x1)
```

## Paso 3 ‚Äî Residuo `%` (par / impar) [pr√°ctica r√°pida]

Calcule si la cantidad de galletas es par o impar usando `%`.
- Si qty % 2 == 0 ‚Üí es par
- Si qty % 2 != 0 ‚Üí es impar

```python
# Complete
es_par_qty_galletas = qty_galletas%2==0
```

```python
print("¬øqty_galletas es par?:", es_par_qty_galletas)
```

## Paso 4 ‚Äî Comparaci√≥n + l√≥gicos (solo evaluar, sin if)

Vamos a crear banderas booleanas:
- aplica_vip = es_vip AND (subtotal >= vip_threshold)
- aplica_cupon = (codigo_cupon == "PROMO5")
- aplica_beneficio = aplica_vip OR aplica_cupon

üëâ Solo calcule las banderas y mu√©strelas.

```python
# Complete
aplica_vip = es_vip and (subtotal >= vip_threshold)
aplica_cupon = (codigo_cupon == "PROMO5")
aplica_beneficio = aplica_vip or aplica_cupon

print("aplica_vip:", aplica_vip)
print("aplica_cupon:", aplica_cupon)
print("aplica_beneficio:", aplica_beneficio)
```

## Paso 5 ‚Äî Total final (aritm√©ticos)

Para mantener este Lab en nivel Secci√≥n 3, NO usaremos if.

Por eso:
- descuento_beneficio = 0 (lo dejaremos en cero en esta secci√≥n)
Luego:
- neto = subtotal - descuento_2x1 - descuento_beneficio
- impuesto = neto * tax_rate
- total = neto + impuesto

```python
descuento_beneficio = 0  # en Secci√≥n 3 lo dejamos fijo

neto = subtotal - descuento_2x1 - descuento_beneficio
impuesto = neto*tax_rate
total = neto+impuesto

print("Neto:", neto)
print("Impuesto:", impuesto)
print("Total:", total)
```

## Paso 6 ‚Äî Exponente `**` (ra√≠z cuadrada)

Defina puntos como:
puntos = int((total ** 0.5) * 10)

```python
puntos = int((total**0.5)*10)
print("Puntos:", puntos)
```

## Paso 7 ‚Äî Ticket (salida simple)

Imprima un ticket con valores redondeados a 2 decimales usando round().

```python
print("===== TICKET =====")
print("Subtotal:", round(subtotal, 2))
print("Descuento 2x1:", round(descuento_2x1, 2))
print("Neto:", round(neto, 2))
print("Impuesto:", round(impuesto, 2))
print("TOTAL:", round(total, 2))
print("Puntos:", puntos)
```

```python
# Verificaci√≥n simple (NO editar)
# Subtotal esperado: 48.0
# Descuento 2x1 esperado: 15.5
# Neto esperado: 32.5
# Impuesto esperado: 4.225
# Total esperado: 36.725

print("\n--- Verificaci√≥n ---")
print("Subtotal ok?:", abs(subtotal - 48.0) < 1e-9)
print("2x1 ok?:", abs(descuento_2x1 - 15.5) < 1e-9)
```
